```python
import datetime 
import random
import numpy as np
import math

population_size = 30;
mutation_rate = 0.01;

# Full Set
gene_set = '''abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ 1234567890, .-;:_!"#%&/()=?@${[]}'''
# Target to reach
target = "collegiateschool"

target_length = len(target)

# function to generate new population for next gen
def new_population(target_length, gene_set, population_size):    
    
    new_pop = [0] * target_length;                          # Initialising the 'population' (password)    
    for i in range(0, population_size): 
        sampled = random.sample(gene_set, target_length)
        new_pop = np.vstack((new_pop, sampled))
    new_pop = np.delete(new_pop, (0), axis = 0)             # Delete the initially padded zeros
    return new_pop;

def fitness(target, samp):                                  # A function to calculate fitness
    fit = 0;
    for i in range(target_length):
        if (str(target[i]) == str(samp[i])):
            fit = fit + 1;
    return fit

# function to find index of top n fitness scores in the new_pop
def Nmaxindex(scores, N):
    finalList = [];
    finalIndex = [];
    
    for i in range(0, N):
       max1 = 0;
       max_index = 0;
       for j in range(len(scores)):
           if scores[j] > max1:
               max1 = scores[j];
               max_index = j;
       np.delete(scores, [max_index]);
       finalIndex.append(max_index);
       finalList.append(max1);
       return finalIndex;
   

# start the main part of the program
# let target_found be the flag to watch for when target is found
generation_count = 1; # count and print the number of generations
target_found = 0;

while(generation_count < 100):
    fit_score = 0; 
    fit_score1 = 0;
    num_of_genes_2mutate = math.ceil(mutation_rate * target_length);
    new_pop = new_population(target_length, gene_set, population_size);    # cross-over done
    
    # check if target achieved, fitness_score = target_length
    # fitness
    for i in range(population_size):
        samp1 = new_pop[i,:];
        fit_scoretemp1 = fitness(target, samp1);
        fit_score1 = np.vstack((fit_score1,fit_scoretemp1));
    fit_score1 = np.delete(fit_score1, (0), axis = 0);
    
    if(max(fit_score1) == target_length):                          # checking for target before mutation
        print("Target found at generation = ", generation_count);
        target_found = 1; # switch flag
    
    # now mutate the children in the locations where target_answer is not yet found
    # in range the last element is not counted
    for j in range(population_size):
        sample = new_pop[j,:];
        for n in range(0, num_of_genes_2mutate):
            randomindex = random.randint(0, len(sample)-1);
            sample[randomindex] = random.choice(gene_set);
        new_pop[j,:] = sample;
                    
    # apply fitness
    for i in range(0, population_size):
        samp = new_pop[i,:];
        fit_scoretemp = fitness(target, samp);
        fit_score = np.vstack((fit_score,fit_scoretemp));
    fit_score = np.delete(fit_score, (0), axis = 0);
    
    if(max(fit_score) == target_length):                         # checking post-mutation
        print("Target found at generation = ", generation_count);
        target_found = 1; # switch flag
    
# Now choose 3 best fitness scores, extract their corresponding children and then form new_pop from them.
    
    arr = np.array(np.transpose(fit_score))         # extract the indices of top 3 scores
    finalArray = np.concatenate(arr.tolist())       # remove array([[3,4,5]]) to array([3,4,5]) and flatten lists 
    top3parents = finalArray.argsort()[-3:][::-1];

    concatenate_parents = [];
    for k in top3parents:
        concatenate_parents = concatenate_parents + new_pop[k,:].tolist(); # putting all elements in best parents in one list (newGeneSet)
    
    arr1 = np.array(np.transpose(fit_score))          # extract the indices of top 3 scores
    final_array1 = np.concatenate(arr1.tolist())      # remove array([[3,4,5]]) to array([3,4,5]) and flatten lists 
    best_parent_index = final_array1.argsort()[-1:][::-1];
    best_parent = new_pop[best_parent_index,:];
    gene_set = concatenate_parents;
    
    best_parent = np.array(best_parent)
    best_parent = best_parent.tolist()    # convert from ndarray to list to print it
    for ii in best_parent:                #  ready to print
        x = ''.join(ii)
    
    del fit_score
    del fit_score1
    del concatenate_parents
    print("\nGeneration = ", generation_count, " ans = ", x)
    generation_count +=1;
```
